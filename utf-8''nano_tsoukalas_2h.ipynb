```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def fun(filepath):
    with open(filepath,'r') as f:
        lines=f.readlines()  
        x=[float(line.split()[0]) for line in lines]
        y=[float(line.split()[1]) for line in lines]
    
    return x,y
    
```

```python
#gia 60 min

x1,y1=fun('/home/nbuser/library/implant_900_60_B.txt')
x2,y2=fun('/home/nbuser/library/implant_950_60_B.txt')
x3,y3=fun('/home/nbuser/library/implant_1000_60_B.txt')



```

```python
plt.figure(figsize=(20,10))
plt.title('Mεταβολές κατανομών στα 60 λεπτά')
plt.semilogy(x1,y1,'b',label="implant_900_60_B")
plt.semilogy(x2,y2,'r',label="implant_950_60_B")    #semilogy for better resolution 
plt.semilogy(x3,y3,'g',label="implant_1000_60_B")
plt.xlabel('Bάθος εισχόρησης στο υπόστρωμα [nm]')
plt.ylabel('Αριθμός των προσμίξεων')
plt.grid()
plt.legend(loc="upper right")
plt.show()
```
